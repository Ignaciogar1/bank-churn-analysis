```python
#Importamos librerías

from matplotlib.ticker import PercentFormatter
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

#Config inicial de gráficos

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
#Cargamos el Dataset

df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
#Creamos columnas derivadas

# 1. Rango de edades
bins_age = [18, 30, 40, 50, 60, 70, 100]
labels_age = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['Age_group'] = pd.cut(df['Age'], bins=bins_age, labels=labels_age, right=False)

# 2. Categoría de saldo
bins_balance = [-1, 0, 50000, 100000, 150000, 200000, df['Balance'].max()]
labels_balance = ['0', '0-50k', '50k-100k', '100k-150k', '150k-200k', '200k+']
df['Balance_category'] = pd.cut(df['Balance'], bins=bins_balance, labels=labels_balance)

# 3. Cliente “high value” (saldo alto y >2 productos)
df['HighValueCustomer'] = ((df['Balance'] > 100000) & (df['NumOfProducts'] >= 3)).astype(int)

# 4. Codificar género para análisis más rápido
df['Gender_code'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Crear columna legible para churn
df['Churn_Label'] = df['Exited'].map({0: 'Se quedó', 1: 'Abandonó'})

# Comprobamos primeras filas
df[['Age', 'Age_group', 'Balance', 'Balance_category', 'NumOfProducts', 'HighValueCustomer', 'Gender', 'Gender_code']].head()
```

```python
#Creamos la base SQLITE local y tabla

conn = sqlite3.connect("churn.db")

df.to_sql("churn", conn, if_exists="replace", index=False)

pd.read_sql("SELECT * FROM churn LIMIT 5;", conn)
```

```python
#consultas SQL analíticas

queries = {
    "Churn_por_pais": """
        SELECT Geography, 
               COUNT(*) AS TotalCustomers, 
               SUM(Exited) AS CustomersExited, 
               ROUND(AVG(Exited)*100, 2) AS ChurnRatePercent
        FROM churn
        GROUP BY Geography
        ORDER BY ChurnRatePercent DESC;
    """,
    "Churn_por_genero": """
        SELECT Gender, 
               COUNT(*) AS TotalCustomers, 
               SUM(Exited) AS CustomersExited, 
               ROUND(AVG(Exited)*100, 2) AS ChurnRatePercent
        FROM churn
        GROUP BY Gender;
    """,
    "Promedio_saldo_salario_por_churn": """
        SELECT Exited, 
               ROUND(AVG(Balance),2) AS AvgBalance, 
               ROUND(AVG(EstimatedSalary),2) AS AvgSalary
        FROM churn
        GROUP BY Exited;
    """,
    "HighValue_exited": """
        SELECT COUNT(*) AS HighValueExited
        FROM churn
        WHERE HighValueCustomer = 1 AND Exited = 1;
    """,
    "Churn_por_rango_edad": """
        SELECT Age_group, 
               COUNT(*) AS TotalCustomers, 
               SUM(Exited) AS CustomersExited, 
               ROUND(AVG(Exited)*100,2) AS ChurnRatePercent
        FROM churn
        GROUP BY Age_group
        ORDER BY Age_group;
    """
}

# Ejecutar cada query y almacenar resultados en un diccionario
results = {}
for name, query in queries.items():
    results[name] = pd.read_sql(query, conn)

# Mostrar resultados
for name, df_result in results.items():
    print(f"\n=== {name} ===")
    print(df_result)
```

```python
palette_main = sns.color_palette("Set2")  # paleta uniforme para todo

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

#Churn por país
sns.barplot(
    x='Geography', 
    y='ChurnRatePercent', 
    data=results['Churn_por_pais'], 
    color=palette_main[0], 
    ax=axes[0,0]
)
axes[0,0].set_title('Churn por país (%)', fontsize=14, fontweight='bold')
axes[0,0].set_ylabel('Churn Rate (%)', fontsize=12)
axes[0,0].set_xlabel('')
axes[0,0].tick_params(axis='x', rotation=20)

#Churn por género
sns.barplot(
    x='Gender', 
    y='ChurnRatePercent', 
    data=results['Churn_por_genero'], 
    color=palette_main[1], 
    ax=axes[0,1]
)
axes[0,1].set_title('Churn por género (%)', fontsize=14, fontweight='bold')
axes[0,1].set_ylabel('')
axes[0,1].set_xlabel('')
axes[0,1].tick_params(axis='x', rotation=0)

#Promedio de saldo y salario según churn
df_avg_melt = results['Promedio_saldo_salario_por_churn'].melt(
    id_vars='Exited', 
    value_vars=['AvgBalance','AvgSalary'], 
    var_name='Variable', 
    value_name='Valor'
)
sns.barplot(
    x='Exited', 
    y='Valor', 
    hue='Variable', 
    data=df_avg_melt, 
    palette=[palette_main[2], palette_main[3]], 
    ax=axes[1,0]
)
axes[1,0].set_title('Promedio Balance y Salario por Churn', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Exited (0=Se quedó, 1=Abandonó)', fontsize=12)
axes[1,0].set_ylabel('Valor promedio', fontsize=12)
axes[1,0].legend(title='Variable', fontsize=10, title_fontsize=12)

#Churn por rango de edad
sns.barplot(
    x='Age_group', 
    y='ChurnRatePercent', 
    data=results['Churn_por_rango_edad'], 
    color=palette_main[4], 
    ax=axes[1,1]
)
axes[1,1].set_title('Churn por grupo de edad (%)', fontsize=14, fontweight='bold')
axes[1,1].set_ylabel('Churn Rate (%)', fontsize=12)
axes[1,1].set_xlabel('Rango de edad', fontsize=12)
axes[1,1].tick_params(axis='x', rotation=30)

plt.show()
```

```python
palette_main = ['lightgreen','salmon']

#High Value vs Churn
hv_ct = pd.crosstab(df['HighValueCustomer'], df['Churn_Label']).reindex([0,1]).fillna(0)
hv_long = hv_ct.reset_index().melt(id_vars='HighValueCustomer', var_name='Churn_Label', value_name='Count')

# Figura 2x2
fig, axes = plt.subplots(2,2, figsize=(12,9))
plt.subplots_adjust(hspace=0.35, wspace=0.25)

sns.barplot(
    x='HighValueCustomer', y='Count', hue='Churn_Label',
    data=hv_long, order=[0,1], hue_order=['Se quedó','Abandonó'],
    palette=palette_main, ax=axes[0,0]
)
axes[0,0].set_title("Clientes High Value: Se quedó vs Abandonó", fontsize=12, fontweight='bold')
axes[0,0].set_xlabel("HighValueCustomer (0=No, 1=Sí)")
axes[0,0].set_ylabel("Número de clientes")
axes[0,0].legend(title="Churn")

#Churn por rango de edad y High Value
df_agehv = df.dropna(subset=['Age_group']).copy()
churn_by_age_hv = df_agehv.groupby(['Age_group','HighValueCustomer'], observed=True)['Exited'].agg(['mean','count']).reset_index()

sns.barplot(
    x='Age_group', y='mean', hue='HighValueCustomer',
    data=churn_by_age_hv,
    hue_order=[0,1],
    palette=palette_main, ax=axes[0,1]
)

handles, labels = axes[0,1].get_legend_handles_labels()
axes[0,1].legend(handles, ['No High Value','High Value'], title="Tipo de cliente")
axes[0,1].set_title("Churn por rango de edad según tipo de cliente", fontsize=12, fontweight='bold')
axes[0,1].set_ylabel("Churn Rate")
axes[0,1].set_xlabel("Grupo de edad")
axes[0,1].set_ylim(0,1)
axes[0,1].yaxis.set_major_formatter(PercentFormatter(1.0))
axes[0,1].tick_params(axis='x', rotation=30)

#Distribución de saldo según churn
sns.histplot(
    data=df, x='Balance', hue='Churn_Label',
    palette=palette_main, kde=True, ax=axes[1,0]
)
axes[1,0].set_title("Distribución de saldo según churn", fontsize=12, fontweight='bold')
axes[1,0].set_xlabel("Balance")
axes[1,0].set_ylabel("Número de clientes")

#Distribución de salario según churn
sns.histplot(
    data=df, x='EstimatedSalary', hue='Churn_Label',
    palette=palette_main, kde=True, ax=axes[1,1]
)
axes[1,1].set_title("Distribución de salario según churn", fontsize=12, fontweight='bold')
axes[1,1].set_xlabel("Salario estimado")
axes[1,1].set_ylabel("Número de clientes")

plt.tight_layout()
plt.show()
```

```python
# Agrupar por país y HighValueCustomer
churn_country_hv = df.groupby(['Geography','HighValueCustomer'])['Exited'].agg(['mean','count']).reset_index()

# Gráfico de barras
plt.figure(figsize=(8,5))
bar = sns.barplot(
    x='Geography', y='mean', hue='HighValueCustomer',
    data=churn_country_hv,
    palette={0:'lightgreen',1:'salmon'},  # asignar colores explícitamente
    hue_order=[0,1]
)

plt.title("Churn por país según tipo de cliente", fontsize=14, fontweight='bold')
plt.ylabel("Churn Rate")
plt.xlabel("País")
plt.ylim(0,1)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1.0))

bar.legend(title="High Value Customer")
plt.show()
```

```python
# Agrupar por rango de edad y HighValueCustomer
churn_age_hv = df.groupby(['Age_group','HighValueCustomer'], observed=True)['Exited'].agg(['mean','count']).reset_index()

# Crear tabla pivot para mostrar porcentajes
churn_age_hv['mean_percent'] = (churn_age_hv['mean']*100).round(2)
churn_age_hv_pivot = churn_age_hv.pivot(index='Age_group', columns='HighValueCustomer', values='mean_percent')
churn_age_hv_pivot.columns = ['No High Value', 'High Value']
print(churn_age_hv_pivot)
```

```python
summary_country_hv = churn_country_hv.copy()
summary_country_hv['ChurnRatePercent'] = (summary_country_hv['mean']*100).round(2)
summary_country_hv = summary_country_hv.pivot(index='Geography', columns='HighValueCustomer', values='ChurnRatePercent')
summary_country_hv.columns = ['No High Value', 'High Value']
print(summary_country_hv)
```
